# Face detection using openCV

```python
# Import the libraries
```

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Read the image to be detected and convert to gray
```

```python
# Read image
face_image = cv2.imread('laliga.jpeg')

# Convert to gray
gray_img = cv2.cvtColor(face_image, cv2.COLOR_BGR2GRAY)
```

```python
# Plot gray image
plt.figure(figsize=(20,10))
plt.imshow(gray_img, cmap='gray')
```

```python
# Load training file
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
# Scale detector
face_detect = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5)

# Print the number of faces found
print("Faces found: {} faces.".format(len(face_detect)))

face_detect
```

```python
#Loop through the faces found
for (x, y, w, h) in face_detect:
    cv2.rectangle(face_image, (x, y), (x+w, y+h), (0, 0, 255), 2)
```

```python
# Plot 
plt.figure(figsize=(20,10))
plt.imshow(cv2.cvtColor(face_image, cv2.COLOR_BGR2RGB))
```
